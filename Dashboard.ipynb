```python
%%writefile app.py
import streamlit as st
import plotly.express as px
import pandas as pd
import numpy as np
from funpymodeling.exploratory import freq_tbl

@st.cache_data
def load_data():
    df1 = pd.read_csv('titanic.csv')
    df2 = pd.read_csv('titanic.csv',index_col='Name')
    table=freq_tbl(df1['Sex'])
    Filtro=table[table['frequency']>=1]
    Filtro_index1=Filtro.set_index('Sex')
    numeric_df1=Filtro_index1.select_dtypes(['float','int'])
    numeric_cols1=numeric_df1.columns
    numeric_df2=df2.select_dtypes(['float','int'])
    numeric_cols2=numeric_df2.columns
    return Filtro_index1,df2,numeric_df1,numeric_cols1,numeric_df2,numeric_cols2
Filtro_index1,df2,numeric_df1,numeric_cols1,numeric_df2,numeric_cols2=load_data()
st.sidebar.title('DASHBOARD')
st.sidebar.header('Sidebar')
st.sidebar.subheader('Panel de selección')
Frames=st.selectbox(label='Frames',options=['Frame 1','Frame 2','Frame 3','Frame 4'])
if Frames=='Frame 1':
    st.title('TITANIC')
    st.header('Frame principal')
    st.subheader('Line Plot')
    check_box=st.sidebar.checkbox('Mostrar tabla')
    if check_box:
        st.write(Filtro_index1)
        st.write(df2)
    Vars_num_selected=st.sidebar.multiselect(label='Variables numericas',options=numeric_cols2)
    figure1=px.line(data_frame=df2,x=df2.index,y=Vars_num_selected)
    st.plotly_chart(figure1)
if Frames=='Frame 2':
    st.title('TITANIC')
    st.header('Frame Secundario')
    st.subheader('Bar Plot')
    Vars_Num=st.selectbox(label='Variables numericas',options=['frequency','percentage','cumulative_perc'])
    if Vars_Num=='frequency':
        numeric_cols1[0]
    if Vars_Num=='percentage':
        numeric_cols1[1]
    if Vars_Num=='cumulative_perc':
        numeric_cols1[2]
    figure2=px.bar(data_frame=Filtro_index1,x=Filtro_index1.index,y=Vars_Num,title=str('Análisis Univariado'))
    figure2.update_xaxes(automargin=True)
    figure2.update_yaxes(automargin=True)
    Button=st.sidebar.button('Mostrar Gráfico tipo BARPLOT')
    if Button:
        st.plotly_chart(figure2)
if Frames=='Frame 3':
    st.title('TITANIC')
    st.header('Frame Terciario')
    st.subheader('Scatter Plot')
    x_selected=st.sidebar.selectbox(label='x',options=numeric_cols2)
    y_selected=st.sidebar.selectbox(label='y',options=numeric_cols2)
    figure3=px.scatter(data_frame=numeric_df2,x=x_selected,y=y_selected,title='Dispersiones')
    st.plotly_chart(figure3)
if Frames=='Frame 4':
    st.title('TITANIC')
    st.header('Frame Cuarto')
    st.subheader('Pie Plot')
    Variable_num=st.sidebar.selectbox(label='Variables numericas',options=numeric_cols2)
    figure4=px.pie(data_frame=df2,names=df2.index,title='Features Passengers',values=Variable_num,width=1600,height=600)
    st.plotly_chart(figure4)
#st.markdown()
```
