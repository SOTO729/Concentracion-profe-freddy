```python
import pandas as pd
```

```python
"""
a) Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” y la columna “Referencia”
b) Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00
c) Todos los registros de “ Tipo: parking, local, industrial y piso”
d) Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”
e) Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”
f) Todos los registros que no sean de “Provincia : Barcelona“
g)Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“
"""
```

```python
data=pd.read_excel('Inmuebles.xlsx',header=1)
```

```python
data
```

```python
#a) Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” y la columna “Referencia”
data=data.drop(columns=['Referencia'])
data=data.drop(0)
data
```

```python
#b) Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00
data_fecha=data[(data['Fecha_Alta']=='2007-04-29 00:00:00')|(data['Fecha_Alta']=='2004-01-02 00:00:00')]
data_fecha
```

```python
#c) Todos los registros de “ Tipo: parking, local, industrial y piso”
data_tipo=data[(data['Tipo']=='Parking')|(data['Tipo']=='Local')|(data['Tipo']=='Industrial')|(data['Tipo']=='Piso')]
data_tipo
```

```python
#d) Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”
data_provincia=data[(data['Provincia']=='Girona')&(data['Superficie']>100)]
data_provincia
```

```python
#e) Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”
data_precio=data[(data['Precio Venta']>1000000)|(data['Vendedor']=='Carmen')|(data['Vendedor']=='Joaquin')]
data_precio
```

```python
#f) Todos los registros que no sean de “Provincia : Barcelona“
data_bar=data[data['Provincia']!='Barcelona']
data_bar
```

```python
#g)Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“
data_oper=data[(data['Operación']!='Alquiler')&(data['Tipo']!='Oficina')]
data_oper
```
