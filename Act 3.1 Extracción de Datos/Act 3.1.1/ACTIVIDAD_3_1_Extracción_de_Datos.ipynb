```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df=pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df.head()
```

```python
df.info()
```

***Filtro basicos para variables cualitativas***

```python
filtro1=df[df['NOMBRE_CLIENTE']=='PÚBLICO EN GENERAL']
filtro1.to_csv('filtro1')
filtro1.head()
```

```python
filtro1=df[df['FECHA_DOC']=='2022-01-17']
filtro1.to_csv('filtro2')
filtro1.head()
```

```python
filtro2=df.iloc[[1,100,1000],:] #Filtro por filas
filtro2.to_csv('filtro3')
filtro2
```

```python
filtro2=df.iloc[:,[7,8,9,10,11,12]] #Filtro por columnas
filtro2.to_csv('filtro4')
filtro2
```

```python
filtro3=df.iloc[10:100,[7,8,9,10,11,12]] #Filtro por filas y columnas
filtro3.to_csv('filtro5')
filtro3.head()
```

```python
df1=pd.read_excel('Detalle precios y productos fabricados 2022.xlsx',index_col=1)
df1.head()
```

```python
filtro3=df1.loc[['2022-01-05','2022-01-17','2022-01-20'],['MARGEN_TOTAL_CALCULADO']] #Filtro por filas vs columnas
filtro3.to_csv('filtro6')
filtro3
```

```python
filtro5=df.tail(90)
filtro5.to_csv('filtro7')
filtro5
```

***Filtros basicos para variables cuantitativas***

```python
filtro6=df[df['COSTO_TOTAL_CALCULADO']>7000]
filtro6.to_csv('filtro8')
filtro6
```

***Filtros basicos con conectores logicos***

```python
filtro7=df[(df['COSTO_TOTAL_CALCULADO']<6000) & (df['NOMBRE_CLIENTE']=='PROVEEDORA DE SEGURIDAD INDUSTRIAL DEL GOLFO')]#AND
filtro7.to_csv('filtro9')
filtro7
```

```python
filtro8=df[(df['CANT']<10) | (df['NOMBRE_VENDEDOR']=='ALFREDO CANELA')]#OR
filtro8.to_csv('filtro10')
filtro8
```

```python
filtro9=df[~(df['PRECIO_UNITARIO']<600) &~ (df['NOMBRE_CLIENTE']=='ANA BARBARA CRUZ PALACIOS')]#NOT
filtro9.to_csv('filtro11')
filtro9
```
