```python
#Probability of failure up to time
def F(t,eta,beta,gamma):
  return 1-np.exp(((t-gamma)/eta)**beta)*100
#Reliability, that is, the probability that failure will not occur up to time
def R(t,eta,beta):
  return np.exp(-(t/eta)**beta)*100
```

```python
weibull=common_failure.copy()
weibull=weibull.drop([6,7]) #Quitar las fallas menos relevantes
weibull['Shape Parameter Beta']=[0.75,4.01,2.25,1.53,0.96,3.93]
weibull['Characteristic life (days) Eta']=[86.7,508.9,195.5,195.8,29.9,169.39]
weibull['Failure type']=['Premature', 'Wear out','Rand + Wear Out','Rand + Wear Out','Random', 'Wear Out']
weibull
```

```python
time=np.linspace(1,365,365)
eq=[]
eq2=[]
for i in range(weibull.shape[0]):
  for j in range(time.shape[0]):
    eq.append(R(time[j],weibull['Characteristic life (days) Eta'][i],weibull['Shape Parameter Beta'][i]))
  eq2.append(eq)
  eq=[]
```

```python
results=pd.DataFrame()
results['Age (Days)']=time
results['Reliability (%)']=eq2[0]
plot=sb.lineplot(data=results,x='Age (Days)',y='Reliability (%)')
plot.set_title(weibull['Failure'][0])
plot.annotate('$beta = 0.75$ \neta = 86.70',xy=(0.8,0.85),xycoords=plot.transAxes)
plt.gcf().set_size_inches(10, 5)
```

```python
def CPUT(t,Cp,Cu):
  eta=169.39
  beta=3.93
  import scipy.integrate as integrate
  def R(t,eta,beta):
    return np.exp(-(t/eta)**beta)
  integral=integrate.quad(R,0,t,args=(eta,beta))
  integral=integral[0]
  return ((Cp*R(t,eta,beta)+Cu*(1-R(t,eta,beta)))/integral)
```

```python
s=[]
time=np.linspace(1,280,280)
for i in range(time.shape[0]):
  s.append(CPUT(time[i],80,800))
```

```python
results=pd.DataFrame()
results['Time (Day)']=time
results['Cost']=s
```

```python
plot=sb.lineplot(data=results,x='Time (Day)',y='Cost',color='black')
plot.set_title(weibull['Failure'][5])
plt.gcf().set_size_inches(10, 5)
```
