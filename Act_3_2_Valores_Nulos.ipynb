```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Valores Nulos**

```python
data=pd.read_csv("inmuebles2.csv",skiprows=1)
data.head()
```

```python
data1=data.drop(['0','Referencia'],axis=1)
data1
```

```python
data1.info()
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
valores_nulos=data1.isnull().sum().sum()
valores_nulos
```

```python
data3=data1.copy()
```

```python
data3['Superficie']=data3['Superficie'].fillna(round(data1['Superficie'].mean(),1))
data3
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3['Precio Venta']=data3['Precio Venta'].fillna(round(data1['Precio Venta'].median(),1))
data3
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3['Tipo']=data3['Tipo'].fillna(method='ffill')
data3
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3['Operación']=data3['Operación'].fillna(method='bfill')
data3
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data3['Fecha Venta']=data3['Fecha Venta'].fillna('2004-06-19 00:00:00')
data3
```

```python
data3['Vendedor']=data3['Vendedor'].fillna(method='ffill')
data3
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data4=data1.copy()
```

```python
data4=data4.dropna()
data4
```
