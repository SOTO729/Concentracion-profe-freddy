***Filtros basicos para variables cuantitativas***

```python
filtro6=df[df['MONTO ADEUDADO']>70000]
filtro6
```

***Filtros basicos con conectores logicos***

```python
filtro7=df[(df['MONTO ADEUDADO']<6000) & (df['FECHA_FACTURA']=='2023-03-14')]#AND
filtro7
```

```python
filtro8=df[(df['MONTO ADEUDADO']<6000) | (df['FECHA_FACTURA']=='2023-03-14')]#OR
filtro8
```

```python
filtro9=df[~(df['MONTO ADEUDADO']<6000) &~ (df['FECHA_FACTURA']=='2023-03-14')]#NOT
filtro9
```

```python
filtro9.to_csv('Entregable1.csv')
```
