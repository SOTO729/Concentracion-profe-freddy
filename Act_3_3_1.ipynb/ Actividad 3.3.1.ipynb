```python
data=data2.copy()
```

```python
cuantitativas=data.iloc[:,1:24]
cualitativas=data.iloc[:,[0]]
```

```python
fig=plt.figure(figsize=(15,8))
cuantitativas.plot(kind='box',vert=False)
plt.title('Valores atÃ­picos del conjunto de datos')
plt.show()
```

```python
y=cuantitativas
Limite_Superior=y.mean()+3*y.std()
Limite_Inferior=y.mean()-3*y.std()
print('Limite superior permitido',Limite_Superior)
print('Limite inferior permitido',Limite_Inferior)
```

```python
data4=cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data4
```

```python
datos_limpios=pd.concat([cualitativas,data4],axis=1)
datos_limpios
```

```python
datos_limpios.to_csv('Ventas_totales_limpio1.csv')
```

```python
y=cuantitativas
percentile25=y.quantile(0.25)
percentile75=y.quantile(0.75)
iqr=percentile75-percentile25
lim_sup=percentile75+1.5*iqr
lim_inf=percentile25-1.5*iqr
print('Limite superior permitido',lim_sup)
print('Limite inferior permitido',lim_inf)
```

```python
data5_iqr=cuantitativas[(y<=lim_sup)&(y>=lim_inf)]
data5_iqr
```

```python
valores_nulos=data5_iqr.isnull().sum()
valores_nulos
```

```python
data4=data5_iqr.copy()
data4=data4.fillna(round(data5_iqr.mean(),1))
data4
```
