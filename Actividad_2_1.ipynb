```python
import pandas as pd
```

```python
data=pd.read_csv('Ventas_totales.csv')
data.head()
```

```python
data.info()
```

```python
data2=data.copy()
```

```python
nulos=data2.isnull().sum().reset_index()
nulos[nulos[0]!=0]
```

```python
data2['salon_ventas']=data2['salon_ventas'].fillna(data2['salon_ventas'].mean())#Rellenar con la media
data2
```

```python
data2['tarjetas_debito']=data2['tarjetas_debito'].fillna(data2['tarjetas_debito'].median())#Rellenar con la media
data2
```

```python
data2['tarjetas_credito']=data2['tarjetas_credito'].fillna(data2['tarjetas_credito'].median())#Rellenar con la media
data2
```

```python
data2['otros_medios']=data2['otros_medios'].fillna(data2['otros_medios'].median())#Rellenar con la media
data2
```

```python
data2['subtotal_ventas_alimentos_bebidas']=data2['subtotal_ventas_alimentos_bebidas'].fillna(data2['subtotal_ventas_alimentos_bebidas'].mean())#Rellenar con la media
data2
```

```python
data2['bebidas']=data2['bebidas'].fillna(data2['bebidas'].mean())#Rellenar con la media
data2
```

```python
data2['almacen']=data2['almacen'].fillna(data2['almacen'].median())#Rellenar con la media
data2
```

```python
data2['panaderia']=data2['panaderia'].fillna(data2['panaderia'].mean())#Rellenar con la media
data2
```

```python
data2['lacteos']=data2['lacteos'].fillna(data2['lacteos'].mean())#Rellenar con la media
data2
```

```python
data2['carnes']=data2['carnes'].fillna(data2['carnes'].mean())#Rellenar con la media
data2
```

```python
data2['verduleria_fruteria']=data2['verduleria_fruteria'].fillna(data2['verduleria_fruteria'].mean())#Rellenar con la media
data2
```

```python
data2['alimentos_preparados_rotiseria']=data2['alimentos_preparados_rotiseria'].fillna(0)#Rellenar con la media
data2
```

```python
data2['indumentaria_calzado_textiles_hogar']=data2['indumentaria_calzado_textiles_hogar'].fillna('NA')#Rellenar con la media
data2
```

```python
data2['electronicos_articulos_hogar']=data2['electronicos_articulos_hogar'].fillna(method='ffill')#Rellenar con la media
data2
```

```python
data2['otros']=data2['otros'].fillna(method='bfill')#Rellenar con la media
data2
```

```python
nulos=data2.isnull().sum()
nulos.sum()
```

```python
data3=data.copy()
data3.dropna(inplace=True)
```
